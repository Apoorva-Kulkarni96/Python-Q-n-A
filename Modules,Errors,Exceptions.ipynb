## Overriding and Import Conflicts

* If two modules have functions with identical names and both are imported using the wildcard *, the latter will override the former.

**Example:**

from moduleA import *

from moduleB import *

* If both moduleA and moduleB have a function named fn, the fn from moduleB will be used. To avoid such conflicts, it's recommended to import modules without using the wildcard and then reference them with the module name.

## Scope and Namespace
* Every module has its own namespace, which is essentially a container of its functions, classes, and variables. When you import a module, you're accessing its namespace.
* By referencing functions with their module name, you can avoid potential naming conflicts. For instance, moduleA.fn() and moduleB.fn() are clear and unambiguous.

#Errors

An error in Python is a deviation from the expected or correct behavior in a program, which can be categorized as syntax errors, logical errors, or runtime errors.

**Syntax Error**:  Occur when the code is not written correctly according to Python's rules. These errors are typically identified by the Python interpreter during the initial parsing of the code, and they prevent the program from running

**Logical Error**: Mistakes in the program's logic. These errors do not result in immediate program crashes or error messages but lead to incorrect or unexpected behavior.

**Runtime Error**: Occur during the execution of the program. They can result from various issues such as dividing by zero, trying to access an undefined variable, or opening a file that doesn't exist.

### Try and Except Blocks

In Python, we use the try-except block to handle exceptions. Its syntax is:

```
try:
    # code that may cause exception
except:
    # code to run when exception occurs
```

Inside the try block, we write the code that may raise an exception.
If an exception occurs, the control of the program jumps to the except block.
If exceptions don't occur, the except block is skipped.

````python
try:
    numerator = int(input("Enter numerator: "))
    denominator = int(input("Enter denominator: "))

    result = numerator/denominator

    print(result)
except:
    print("Denominator cannot be 0. Try again.")

print("Program ends")
````

**Argument of an Exception**

An exception can have an argument, which is a value that gives additional information about the problem that caused the exception. The contents of argument vary by exception.

You can access the exception's argument using the `as` clause when catching the exception in a `try` and `except` block. For example:

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"An error occurred: {e}")
```

In this code, the exception `e` contains the argument, which is the denominator causing the `ZeroDivisionError`.

````python
l1 = [2,0,"why",None]
for x in l1:
    try:
        print("Current Element -", x)
        result = 5/x
        print("Result -", result)
    except ZeroDivisionError:
        result = 5
        print("Result - ", result)
    except Exception as e:
        print("Exception -", e)

    print("-"*50)
````

### Else and Finally Blocks

You can use a `try` and `except` block in combination with `else` and `finally` blocks.

* The code inside the `try` block is where you place the code that may raise an exception.
* The `except` block is where you handle the exception if one occurs.
* The `else` block contains code that executes if no exceptions are raised in the `try` block.
* The code inside the `finally` block is always executed, whether an exception occurs or not.

```python
try:
    # Code that may raise an exception
except SomeException as e:
    # Handle the exception
else:
    # Code that executes if no exception is raised
finally:
    # The `finally` block is separate and contains code that always executes, e.g., cleanup or resource release.
```

The `else` block is useful for performing actions when no exceptions occur, while the `finally` block is typically used for ensuring that certain actions, like resource cleanup or file closure, always take place, regardless of whether an exception occurred or not.

````python
try:
  i  = int(input())
  print('try')
except:
  print("except") #if there is error in try
else:
  print('else') #if there is no error in try
finally:
  print('finally')  # always run
````

### Issuing Warnings

We may want to issue a warning instead of silently continuing. For this, the `warnings` module from the standard library is useful. We use the `warnings.warn()` method to issue the warning.

````python
import warnings
````

````python
def divide(a, b):
    if b == 0:
        # Issue a warning instead of raising an exception
        warnings.warn("Division by zero is not recommended", category=RuntimeWarning)
    return a / b

result = divide(10, 0)

# By default, warnings are not shown; you can control the behavior of warnings using the -W command line switch or the warnings module settings.
````
