```python
df=pd.read_csv("tips.csv")
```

```python
df1=df.groupby('day')['tip'].mean()
df1.plot(kind='bar',color='Orange')
plt.show()
```

```python
df2=df.groupby('day')[['total_bill','tip']].mean().reset_index()
df2
```

```python
df2.plot(kind='bar', stacked =True)
plt.show()
```

```python
df3=df.groupby(['day', 'smoker'])['tip'].mean()
df3.plot(kind='bar')
plt.show()
```

```python
df3=df.groupby(['day', 'smoker'])['tip'].mean().unstack()
df3.plot(kind='bar')
plt.show()
```

```python
df4=df.groupby(["time"])['tip'].mean()
df4.plot(kind='bar')
plt.show()
```

```python
colors = {
    'Male': 'darkorange',
    'Female': 'darkblue'
}
```

```python
for gender in df['sex'].unique():
    gender_data=df[df['sex']==gender]
    plt.scatter(gender_data['total_bill'],gender_data['tip'], color=colors[gender], label=gender,alpha=0.4)
    plt.legend()
plt.show()
```
